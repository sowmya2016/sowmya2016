2.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

To find the proportion of gunny bags with a breaking strength of less than 3.17 kg per sq cm in a normal distribution, we can use the standard normal distribution (z) table. The formula to convert a raw score (X) to a z-score is given by:

z = (x - mean)/std dev

z = (3.17-5)/1.5

P(z < -1.22) =0.1112

So the proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm  is 0.1112

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Given data
mean = 5
std_dev = 1.5
X = 3.17

# Calculate the z-score
z = (X - mean) / std_dev

# Generate data points for CDF plot
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y_values = norm.cdf(x_values, mean, std_dev)

# Plot the CDF
plt.plot(x_values, y_values, label='CDF')
plt.axvline(x=X, color='r', linestyle='--', label=f'X = {X}')
plt.axhline(y=norm.cdf(X, mean, std_dev), color='r', linestyle='--', label=f'P(X < {X:.2f})')

# Highlight the point on the X-axis
plt.scatter(X, norm.cdf(X, mean, std_dev), color='r')

# Add labels and title
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Cumulative Probability')
plt.legend()

# Show the plot
plt.show()
````

3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

H0 : Stone found to be of adequate hardness >=150

Ha : Unpolished stone is not suitable for printing < 150


````python
import numpy as np
from scipy.stats import ttest_1samp
df = pd.read_csv('Zingaro_Company.csv')
# Sample data (hardness measurements)
stone_hardness = df['Unpolished '].tolist()
# Null hypothesis: Stone hardness is >= 150
# Alternative hypothesis: Unpolished stone is not suitable for printing (mean hardness < 150)

# Define the critical value (alpha level)
alpha = 0.05

# Perform one-sample t-test
t_statistic, p_value = ttest_1samp(stone_hardness, 150)

# Output the results
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')

# Compare p-value with alpha to decide whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis. There is enough evidence to support the alternative hypothesis.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.")
````

````python
import pandas as pd
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Read data from your CSV file or other formats
data = pd.read_excel('Dental+Hardness+data.xlsx')
print(data.columns)
print(data.head())

alloy_type_1_data = data[data['Alloy'] == '1']['Response']
alloy_type_2_data = data[data['Alloy'] == '2']['Response']
````

````python
# Conduct ANOVA test
f_statistic, p_value = f_oneway(alloy_type_1_data, alloy_type_2_data)

# Write down conclusions from the test results
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference in implant hardness among dentists.")

    # Conduct Tukey's HSD test for post hoc analysis
    tukey_results = pairwise_tukeyhsd(data['Response'], data['Dentist'])
    print(tukey_results)

    # Identify for which pairs the implant hardness differs
    significant_pairs = tukey_results.reject

    print("Pairs with significant differences:")
    for pair in significant_pairs:
        print(pair)
else:
    print("Fail to reject the null hypothesis. There is no significant difference in implant hardness among dentists.")
````

**Problem ** **4.2**

Null hypothesis: There is no significant difference in implant hardness among methods for each type of alloy.

Alternate hypothesis: There is a significant difference in implant hardness among methods for each type of alloy.

````python
import pandas as pd
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Assuming you have a DataFrame named 'data' with columns: Dentist, Method, Alloy, Temp, Response


# Separate data for each type of alloy
alloy_type_1_data = data[data['Alloy'] == '1']
alloy_type_2_data = data[data['Alloy'] == '2']

# State the null and alternate hypotheses
# Null hypothesis: There is no significant difference in implant hardness among methods for each type of alloy.
# Alternate hypothesis: There is a significant difference in implant hardness among methods for each type of alloy.

# Check assumptions of the hypothesis test (Normality and Homogeneity of variances)
# You may use graphical methods and statistical tests for this, depending on your data distribution.

# Conduct ANOVA test
f_statistic, p_value = f_oneway(alloy_type_1_data['Response'], alloy_type_2_data['Response'])

# Write down conclusions from the test results
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference in implant hardness among methods.")

    # Conduct Tukey's HSD test for post hoc analysis
    tukey_results = pairwise_tukeyhsd(data['Response'], data['Method'])
    print(tukey_results)

    # Identify for which pairs the implant hardness differs
    significant_pairs = tukey_results.reject

    print("Pairs with significant differences:")
    for pair in significant_pairs:
        print(pair)
else:
    print("Fail to reject the null hypothesis. There is no significant difference in implant hardness among methods.")
````

4.3 What is the interaction effect between the dentist and method on the hardness of dental implants for each type of alloy?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is stored in a variable called 'df'
# If not, replace 'df' with the actual variable name holding your dataset

# Example code:
df = pd.read_excel("Dental+Hardness+data.xlsx")
#data = pd.read_excel('Dental+Hardness+data.xlsx')

# Set the style of the visualization
sns.set(style="whitegrid")

# Iterate over each alloy type and create an interaction plot
for alloy_type in df['Alloy'].unique():
    plt.figure(figsize=(10, 6))

    # Filter data for the current alloy type
    alloy_data = df[df['Alloy'] == alloy_type]

    # Create an interaction plot using seaborn
    interaction_plot = sns.lineplot(x='Dentist', y='Response', hue='Method', data=alloy_data, marker='o', err_style="bars")

    # Set plot title and labels
    plt.title(f'Interaction Plot for Alloy Type {alloy_type}')
    plt.xlabel('Dentist')
    plt.ylabel('Response')

    # Show the legend
    plt.legend(title='Method')

    # Show the plot
    plt.show()
````

````python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Assuming your dataset is stored in a variable called 'df'
# If not, replace 'df' with the actual variable name holding your dataset

# Example code:
df = pd.read_excel("Dental+Hardness+data.xlsx")

# Fit an ANOVA model for each alloy type
for alloy_type in df['Alloy'].unique():
    alloy_data = df[df['Alloy'] == alloy_type]

    # Check for missing values
    if alloy_data.isnull().sum().any():
        print(f"Warning: Missing values in the dataset for Alloy Type {alloy_type}.")
        continue

    # Check for sufficient data
    if len(alloy_data) == 0:
        print(f"Warning: No data for Alloy Type {alloy_type}.")
        continue

    # Fit the ANOVA model
    model = ols('Response ~ Dentist * Method', data=alloy_data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # Print the ANOVA table
    print(f"ANOVA for Alloy Type {alloy_type}:\n")
    print(anova_table)
    print("\n")
````

1. ANOVA Results for Alloy Type 1:
The ANOVA for Alloy Type 1 suggests that there is a significant interaction effect between the dentist and method on the hardness of dental implants (p < 0.05).
This implies that the effect of the dentist on hardness depends on the method used, and vice versa.
2. ANOVA Results for Alloy Type 2:
The ANOVA for Alloy Type 2 also indicates a significant interaction effect between the dentist and method on hardness (p < 0.05).
Similar to Alloy Type 1, the dentist's effect on hardness is influenced by the method, and vice versa.
3. Overall Trends:
Across both alloy types, it appears that the combination of the dentist and method has a significant impact on the hardness of dental implants.
It's important to note that the p-value for the interaction effect is a measure of statistical significance. The actual effect size and practical significance should also be considered.
